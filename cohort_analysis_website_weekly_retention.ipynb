```python
df['date']=pd.to_datetime(df.accessed_date).dt.date   
df.head()
```

```python
df['first_open']=df.groupby('ip', as_index=False)['date'].transform('min')
```

```python
df
```

```python
days_diff=df.date-df.first_open
```

```python
days_diff.dt.days+1
```

```python
df['cohort_index']=days_diff.dt.days+1
```

```python
df
```

```python
df.cohort_index.unique()
```

```python
df.first_open=pd.to_datetime(df.first_open)
```

```python
df.dtypes
```

```python
cohort_count=df.groupby(['first_open', 'cohort_index'], as_index=False).agg({'ip':pd.Series.nunique})
cohort_count
```

```python
cohort_count=cohort_count.pivot_table(index='first_open', columns='cohort_index', values='ip')
cohort_count
```

```python
cohort_size=cohort_count.iloc[:, 0] #df.groupby('first_open', as_index=False).agg({'ip':'nunique'})
cohort_size
```

```python
retention = cohort_count.divide(cohort_size, axis=0)
retention.round(2)
```
